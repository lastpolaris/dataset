```python
import os
import os.path as osp
import json 
```

```python
def write_json_data(dict,json_path):
    with open(json_path,'w') as r:
        json.dump(dict,r)
    r.close()
```

```python
file = open("list.txt") # 存放有数据集所在的文件夹
dataset = file.readlines()
dict = {}
for data in dataset:
    view = []
    data = data.replace('\n','')
    viewnames = os.listdir(data)
   
    data_dict = {}
    for viewname in viewnames:
        imgs = os.listdir(osp.join(data,viewname,'img'))
        print(imgs)
        with open(osp.join(data,viewname,viewname+'.json')) as metajson:
            meta = json.load(metajson)
            print(meta)
        with open(osp.join(data,viewname,'groundtruth_rect.txt')) as gtfile:
            ground_truth = gtfile.readlines()
            for i in range(len(ground_truth)):
                ground_truth[i] = ground_truth[i].replace('\n','').split(', ')
                ground_truth[i] = [int(ground_truth[i][0]),int(ground_truth[i][1]),int(ground_truth[i][2]),int(ground_truth[i][3])]
            print(ground_truth)
        view_dict = {
            'imgs': [osp.join(data,viewname,'img',img).replace('\\','/')for img in imgs],
            'meta':meta,
            'gt_rect':ground_truth
        }
        data_dict[viewname] = view_dict
    dict[data] = data_dict
    write_json_data(dict,'MultiViewCAR.json')  # 生成一个数据集信息的json文件
file.close()
    
                    


```
